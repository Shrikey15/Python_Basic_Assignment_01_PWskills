```python
import numpy as np
import statistics
'''
Q2. What are the measures of central tendency, and when should you use each? Discuss the mean, median,
and mode with examples and situations where each is appropriate.
Answer:
    The measures of central tendency describe the center or typical value in a dataset.
    The three main measures are, Mean, Median, Mode, each is used in different situations depending on the
    data type, distribution and any potential outliers.

1.Mean(Average):
    The mean is the sum of all values divided by the number of values.
    It is best used for continuous data without extreme outliers, as outliers can skew the data
Example:
        In a data set of delivery time: [25, 32, 36, 43, 39] the mean is :
            (25, 32, 36, 43, 39)/5 = 35
'''
# Delivery time dataset with no outliers
delivery_time = np.array([25, 32, 36, 43, 39])

# Calculating mean
mean_delivery = np.mean(delivery_time)
print('Average delivery time is:\n', mean_delivery, "minutes")
'''
When to use:
    Use the mean for normally distributed data where all values are relatively close to each other,
    for instance Calculating the average delivery time of food home delivery.

2.Median(Middle value/ Values)
    The median is the middle value in an ordered dataset, if there is an even number of values, it is the
    average of the two (physical) middle values.
Example:
    In this dataset [5, 15, 20, 25, 100], the median is 20, the median is better tha the mean, as the
    mean would be skewed by the outlier 100.
'''
print()
# dataset with outlier
data = np.array([5, 15, 20, 25, 100])

# calculating the mode
median = np.median(data)
print("The median of the data is: \n", median)
'''
When to use:
    Use the median when the data has outliers or is skewed, like household incomes or property prices

3.Mode:
    The mode is the most frequently occurring values(s) in a dataset.
    A dataset can have multiple modes (Bimodal, Multimodal) if there are several values with the highest frequency.
Example:
    In a dataset representing shoe sized [8,8, 9, 10, 10, 10, 11] then the mode is 10, which has the highest frequency.
'''
print()
# Creating data array for calculating mode
shoes_size = np.array([8,8, 9, 10, 10, 10, 11])

# calculating the mode of the data
mode = statistics.mode(shoes_size)
print("The mode of the following data is: \n",mode)
'''
When to use mode:
    The mode is particularly useful for categorical data or when identifying the most common occurrence, like finding
    the most popular product sold

insights:
    Choosing the Right Measure:
        1.Mean is suitable for continuous data with a normal distribution.
        2.Median works well for skewed data or when dealing with the outliers.
        3.Mode is ideal for categorical data or when identifying the most frequent occurrence is useful
    '''
print()

```

```python
"""
3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
Answer:
    Dispersion describes how spread out data points are withing a dataset.
    It shows how much data values deviate from the mean, which helps to understand variability.
    Variance and Standard deviation are two main measures of dispersion that helps quantify this spread.

1.Variance:
    Variance calculates the average squared deviation of deviation of each data point from the mean.
    A higher variance indicated greater spread, meaning data points are far from the mean, while a low variance means
    they are close .

1.Standard Deviation:
    Standard deviation is the square root of variance, returning the spread in the original units of the data, making
    it more interpretable, higher values suggest wider dispersion.

Example:
# Suppose we have data on students scores
# [35, 36, 40, 50, 60]
"""
'''
    1.Variance gives an average of the squared deviations from the mean scores.
        if it is high, scores are widely scattered and if low, scores are close to each other

    Code_Example: Calculating variance.
'''
# sample data
scores = [35, 36, 40, 50, 60]
print("scores:\n", scores)

print()
# Calculating mean
mean_scores = np.mean(scores)
# Calculating varaince
variance = round(np.var(scores))

# Printing results
print("The mean score is:\n", mean_scores)
print()
print("The variance is:\n", variance)

'''
    2.Standard deviation (taking the square root of variance ) will provide a clearer, scaled measure of how scores
    deviate from the mean.

    Example: calculating standard deviation from the above sample data.
'''
print()
# data
# scores = scores

# Calculating standard devaition
std_dev = round(np.std(scores), 2)


# Printing the standard deviation result
print("The standard deviaiton of the square root of variance is:\n", std_dev)

'''
When to use each:
    Variance :
    Use variance when you are interested in understaning consistency across a dataset with more precise units, like
    squared terms in scientific data.

    Standard_deviation:
    use standard deviation in everyday scenarios for a straight forward sense of how much fluctuation there is around
    the mean.
    This is especially helpful in finance, research and quality control to interpret spread directly in the data
    original units

insights:
    1.A variance 86.0 indicates that scores are moderately spread around the mean(82.0)
    2.A standard deviation of 9.27 means most scores lie within the 9.27 of the mean providing a clearer, scaled view
    of this spread
'''
print()
```

```python
"""
Q5. Discuss the role of random sampling in making inferences about populations.
Answer:
    Ramdom sampling plays a crucial role in ensuring that inferences made about a populations are reliable,
    unbiased and representative.
    It helps us make generalizations about the entire population based on small randomly selected sample.

    Roles of Random Sampling in laking inferences about populations are as following

1.Represents the whole populations: By giving each individual in the population an equal chance of being selected,
    random sampling ensures that the sample is a fair representation of the whole population.
    This enables us to make accurate estimated about populations parameters , such as the average age, height or income.

Example:
    If we want to estimate the average age of all the students in a school, randomly selecting a subset of students
    from different grades ensures that we have a fair representation of the entire student body

2.Reduces bias:
    Random sampling eliminated the risk of selection bias, which occurs when certain individuals or groups are
    more likely to be included in the sample than others.
    Without random sampling , the results may not reflect the true diversity of the population, leading to
    misleading conclusions

Example:
    If we only survey students from a single class, the results will not represent the diverse range of students in the
    school,
    Random sampling prevents this.

3.Enables Statistical Inteferences :
    Random sampling allows you to apply statistical techniques, like calculating  confidence intervals and performing,
    hypothesis tests, which help us make informed inferences about the population.
    Theses techniques assume the sample is random and representative

Example:
    After randomly sampling 100 people from a city, we can estimate the average income of the entire city, along with
     a margin of error, based on our sample

Code example of Random sampling and statistical interference:
    Lets consider a population of 1000 people, each with an age randomly chosen between 20 and 70, we will
    randomly sample 100 people from this population, estimate the sample mean and calculate the confidence interval for
    the population mean.
"""
# creating a population of 100 people with ages ranging from 20 to 80
population = [random.randint(20, 70,) for i in range(1000)]

# Randomly sample 100 people form the population
sample_size = 100
sample = random.sample(population, sample_size)

# calculate the average age of the sample
average_age_sample = sum(sample) / len(sample)
print("Average age of the sample: ", average_age_sample)

# Calculating the average age of the entire population
average_age_population = sum(population) / len(population)
print("Average age of the population: ", average_age_population)

# Sample mean and confidence interval
sample_mean = np.mean(sample)
std_error = np.std(sample, ddof= 1)/ np.sqrt(100)
confidence_interval = stats.t.interval(0.95, 99, sample_mean, std_error)

# printing sample mean
print(f"Sample mean:{sample_mean}")

# printing condfidence interval
print(f"95% Confidence Interval of sample mean: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# printing standard error
print(f"Standard error: {std_error:.2f}")


'''
insights
Random sampling not only helps us make accurate inferences about a population but also quantifies the uncertainly in
these estimates through techniques like confidence intervals.
It gives us a  solid foundation to make generalizations while minimizing bias and errors.
'''
print()
```

```python
'''
Q7. What is the inter quartile range (IQR), and how is it used to detect outliers?
Answer:
    The inter quartile range (IQR) is a measure of how spread out the middle 50% of data is,
    it helps us understand the range where most of the data lies

Q1.The first quartile, or 25%th percentile which marks the point below which 25% of the data lies

Q3.The third quartile, or 75th percentile which marks the point below which 75% of the data lies

The IQR is simply the difference between Q3 and Q1

IQR = Q3-Q1

How does it detects the outliers?
    to find outliers, we calculate two limits, it is considered an outliers.

Example:
    lets say we have a small dataset:
    [10, 12, 15, 18, 20, 25, 30, 35, 40, 45]

Step_1: Find the Q1, Q3 and IQR.
    Q1 = 15
    Q3 = 35
    IQR - 35 - 15 = 20

Step_2: Find the limits.
    lower_bound = 15 - (1.5 x 20) = -15
    Upper_bound = 35 + (1.5 x 20 = 65

Step_3
    Check points below -15 or above 65 are outliers.

    In this case, all the data points are between -15 and 65, so there are no outliers

Code_Example:
'''
data = [10, 12, 15, 18, 20, 25, 30, 35, 40, 450]
print("data\n", data)

print()
# finding Q1, Q3 and IQR
Q1 = np.percentile(data, 25)
print("First Quartile:", Q1)
print()
Q3 = np.percentile(data, 75)
print("Third Quartile:", Q3)
print()
IQR = Q3 - Q1
print("IQR of first and third quartile:", IQR)

# Calculate the limits
print()
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + (1.5 * IQR)


# Detect the outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"lower_bound: {lower_bound}")
print()
print(f"upper_bound: {upper_bound}")
print()
print(f"Outliers: {outliers}")

'''
Explanation:
    1.IQR measures how spread out the middle 50% of your data is
    2.By using Q1 and Q3, we can calculate the IQR
    3.We use IQR to find outliers, which are values that lie outside the limits based on 1.5 times the IQR

Insights:
    The Interquartile range (IQR) is a powerful and simple tool for detection outliers in a dataset, unlike other
    methods, such as using the mean and standard deviation, IQR is not affected by extreme values (skewed data).
    This makes IQR especially useful when you are working with data might have a few unusually high or low values
    that could distort other calculations
    '''
print()
```

```python
'''
Q.8 Discuss the conditions under which the binomial distribution is used.
Answer:
    The binomial distribution is a probability distribution that models scenarios where there are exactly two possible
    outcomes: Success and Failure, To use binomial the conditions are as following.

1.Fixed number oof Trails:
    The experiment is repeated a fixed number of times. for example flipping a coin 10 times

Example:
    Suppose you flip a coin 10 times, here the total number flips is fixed which is 10,
    if you want to calculate the probability of getting exactly 6 heads in these 10 flips,
    you can use binomial distribution.
Code_Example:
'''
#number of flips
n = 10
# probability
p = 0.5
probability = binom.pmf(6, n, p)
print(f"Probability of getting exactly 6 heads out of 10: {probability:.5f}")
'''
2.Two possible outcomes:
    Each trial must result in a success or failure,
    for example rolling a 6 on a die (success) or not rolling a 6 (failure)
Example:
    For each coin flip, there are only two outcomes, heads or tails.
    This success or failure nature is ideal for a binomial distribution, heads can be considered a success and tails
    as failure.
for example if you want to calculate the probability of exactly 5 out of 10 people choosing Tea given that there
is a 60% chance each person prefers tea, you can use binomial distribution.
Code_Example:
'''
# number of people
n = 10
p = 0.6 #Probability of choosing tea

probability = binom.pmf(5, n, p)
print(f"Probability of exactly 5 out of 10 people choosing tea: {probability:.4f}")

'''
3.Constant Probability:
      The probability of success is the same in evey trial,
      for example, the probability of drawing a red card from a well shuffled deck remains constant each
      time (assuming replacement).
Example:
    consider a basketball player who has 60% success rate for free throws, if they take 15 shots, the probability of
    scoring on each shot remains constant at 0.6, you can calculate the probability of making exactly 10 shots
    with this approach.
Code_Example:
'''
#Total shots
n = 15
# success rate
probability = binom.pmf(10, n, p)
print(f"Probability of making exactly 10 shots out of 15 is: {probability:.4}")
'''
4.Independent between trials:
    The outcome of one trial does not affect others,
    for example, tossing a coin multiple times.
Example:
    You toss a coin 8 times , each toss is independent of the others, meaning the result of one toss does not affect
    the next
    If you want to calculate the probability of getting exactly 3 heads out of 8 tosses, given each toss has 50% chance
    of landing heads, you can apply the binomial distribution
Code_Example
'''
#number of tosses
n = 8

#probability for 3 times head
p = 0.5
probability = binom.pmf(3, n, p)
print(f"Probability of getting exactly 3 heads out of 8 tosses: {probability:.4f}")

'''
Insights:
    In this example, the binomial distribution allows us to estimate the probability of a specific outcome across
    multiple Trials.
    It is particularly useful when the event has a fixed probability of occurring ( like detecting a detect).
    In real world settings, such probabilities help manufactures estimate when adjustments are needed in the
    production process.
    By running simulations or adjusting the parameter, we can explore different scenarios, which makes the binomial
    distribution valuable in decision making where outcomes and probabilities play a key role.
    '''
print()
```

```python
"""
Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
Answer:
    The normal distribution is a continuous probability distribution that is symmetrical and bell shaped, centered
    around its mean
    It has several key properties.
1.Symmetrical :
    The distribution is symmetric around the mean, meaning values are equally like to fall on either side.

2.Mean, Median, mode are Equal:
    In a perfect normal distribution, the mean median and mode all fall at the center of the curve

3.Ball Shape:
    Most data points cluster around the mean, and the frequency decreases as you move further from the mean,
    forming a bell shaped curve.

4.Defined by Mean and Standard Deviation:
    The shape of a normal distribution curve is determined by its mean(center) and standard deviation (spread)


01.The Empirical Rule (68-95-99.7)
    The empirical rule gives an approximation for the spread of data in a normal distribution

    1.68% of data points fall within 1 standard deviations from the mean.

    2.95% of the data points fall within 2 standard deviations from the mean

    3.99.7% of the data points fall within 3 standard deviations from the mean

This rule helps estimate probabilities and understand data spread for normally distributed datasets.

Example:
    if the average test score in a class is 70 with a standard deviation of 5, then
    1.About 68% of the students scored between 65 and 75
    2.About 95% scored between 60 and 80
    3.About 99.7% scored between 55 and 85
The empirical rule is a quick way to interpret data, specially in field like quality control, finance and
social sciences.

Code_Example:
    """


# Paras for the normal distribution
mean = 0
std_dev = 1

# Generate data for the normal distribution
x = np.linspace(-5, 5, 1000)  # X-axis values
y = stats.norm.pdf(x, mean, std_dev)  # Y-axis values for normal distribution

# Plotting the normal distribution curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Normal Distribution", color="blue")

#Mean, Median, and Mode
plt.axvline(mean, color='red', linestyle='--', label="Mean = Median = Mode")

# Displaying the plot
plt.title("Normal Distribution with Symmetry, Bell Shape and Mean, Median, Mode Equality")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)


'''
Explanation:
    1.Symmetry and belll shape:
    The plot create by stats.norm.pdf(x, mean, std_dev) is symmetric around the mean(0) and has a bell
    shaped curve.

    2.Mean, median, mode:
        In a normal distribution, the mean and mode are equal, this is shown by the vertical red dashed line at
        x = 0, which represents the mean of the distribution

    3.Defined by mean and standard deviation:
        The mean and std_dev variables define the characteristics of the distribution, changing them will shift the
        center or adjust the width of the bell curved.

Insights:
    The visual representation helps you understand that a normal distribution is not just a theoretical concept
    but can be easily modeled and analyzed using simple code, The symmetry of the curve shows how most of the
    data is clustered around the mean with the probability of extreme values dropping significantly as we move
    away from the center
'''
plt.show()
```

```python

'''
Q10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.
Answer:
    Imagine you are standing at a bus stop, the busses arrive randomly, but on average 2 busses arrive every hour,
    The number of busses arrive every hour, the number of busses arriving each hour can be modeled as
    a poisson process because

    1.The busses arrive independently
    2.The average number of busses per hour remains constant
    3.The arrivals happen randomly over time, but the average rate is predictable
Code_Example:
'''
# Average number of buses per hour
rate = 2

# We want the probability of 5
k = 5

# poisson calculation with formula
probability = (rate ** k * math.exp(-rate)) / math.factorial(k)

print(f"Probability of exactly 5 busses arriving in an hour is: {probability:.4f}")
'''
Explanation:
    1.2 on an average, 2 busses arrive every hour
    2.We are calculating the probability of exactly 3 buses arriving in the next hour.
    3.the formula calculates this probability based on the average arrival rate and the number of busses we are
    interested in.

Insights:
    This poisson helps us estimate how likely a certain number of busses will arrive at the bus stop within a specific
    time frame, given the average arrival rate, this concept can be extended to other real life scenarios, like
    the number of emails received in an hour, the number of customers entering a store, or even the number
    of accidents happening on a road understanding this process allows business and services to better manage
    resources, plan schedules and optimize operations
'''
print()
```

```python
'''
Q12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.
Answer:
    consider a data representing the number of books read by individuals and their reading satisfaction score
    (on a scale of 1 to 10)
Person        Books Read        Satisfaction
1             1                 4
2             2                 5
3             3                 6
4             4                 7
5             5                 8

Calculations of covariance and correlation:
    1.Covariance measures how two variables change together, if both variable increases or decreases, the covariance
    will be, Positive.
    If one increases and the other decreases, the covariance will be negative.
Code_Example:Covariance
'''
print("Example of books readed and its satisfaction rating in above data set")
# Data of Books read and Satisfaction rating
books_read = np.array([1, 2, 3, 4, 5])
satisfaction = np.array([4, 5, 6, 7, 8])

# Calculating Covariance
covariance = np.cov(books_read, satisfaction)[0, 1]

print(f"Covaraince: {covariance}")
'''
    2.Correlation standardizes the covariance to show the strength and direction of the relationship.
    The correlation value ranges from -1 to 1

        1. 1 indicates a perfect positive correlation
        2. -1 indicated a perfect negative correlation
        3. 0 indicates no relationship
Code_Example: Correlation
'''
# calculating Correlation
correlation = np.corrcoef(books_read, satisfaction)[0, 1]
print(f"Correlation: {round(float(correlation), 4)}")
'''
Explanation:
    Covariance:
        (2.5) This positive value indicated that as the number of books read increases, the satisfaction score also
        increases, both variables move in the same direction

    Correlation:
    (1.0)This positive value of 1 shows a perfect positive linear relationship between the two variables, as
    the number of books read increases, the satisfaction score increases consistently.

Insights:
    1.covariance tells us the direction of the relationship but not its strength.

    2.Correlation is a standardized measure that shows how string the relationship is,
        it is a better measure when comparing different datasets, as it is always between -1 and 1

    '''
print()
```
