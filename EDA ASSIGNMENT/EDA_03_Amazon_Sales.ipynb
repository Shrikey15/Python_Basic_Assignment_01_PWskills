```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
dataframe = pd.read_csv("amazon.csv")
```

```python
df = dataframe
```

```python
df.columns
```

```python
df
```

```python
# 1.What is the average rating for each product category?

df.rating.unique()
df.rating = df.rating.replace("|", np.nan)
df.rating = df.rating.astype(float)
df.rating.dtype
average_rating_for_each_category = df.groupby("category")["rating"].mean().reset_index()
average_rating_for_each_category.columns = ["category", "average_rating"]
average_rating_for_each_category
```

```python
# 2.What are the top rating_count products by category
top_ratingcount_by_category = df.sort_values(by=["category", "rating_count"],ascending=False)
top_ratingcount_by_category = top_ratingcount_by_category.groupby("category").head(1).reset_index(drop=True)
top_ratingcount_by_category
```

```python
# 3. What is the distribution of discounted prices vs. actual prices

df.discounted_price.dtype
df.discounted_price.unique()
df.discounted_price =  df.discounted_price.str.replace("‚Çπ", "")
df.discounted_price = pd.to_numeric(df.discounted_price , errors="coerce")
df.discounted_price.dtype
# converted discounted_price to numeric

# convertin actual to an numeric column
df.actual_price = df.actual_price.astype(str)
df.actual_price = df.actual_price.replace("‚Çπ", "", regex=True)
df.actual_price = df.actual_price.replace(",", "", regex=True)
df.actual_price = df.actual_price.astype(float)

print("Discounted Price Distribution information.\n",round(df.discounted_price.describe(),2))
print("Actual Price Distribution information",round(df.actual_price.describe(),2))

# making plot for visualising the Distribution
plt.figure(figsize=(11, 6))
plt.subplot(1,2,1)
sns.histplot(df.discounted_price, kde=True)
plt.title("Histogram of Discounted Price Distribution", fontweight="bold")
plt.xlabel("Discounted Price")

plt.subplot(1,2,2)
sns.histplot(df.actual_price, kde=True)
plt.title("Histogram of Actual Price Distribution", fontweight="bold")
plt.xlabel("Actual Price")

plt.tight_layout()
plt.show()

```

```python
# 4.How does the average discount percentage vary across categories
df.discount_percentage.dtype
df.discount_percentage.unique()
df.discount_percentage = df.discount_percentage.str.replace("%", "")
df.discount_percentage = df.discount_percentage.astype(int)

df.groupby("category")['discount_percentage'].mean().reset_index()
```

```python
# 5. What are the most popular product names
popular_product_name = df.product_name.value_counts().reset_index()
popular_product_name
```

```python
# 6. What are the most popular product keywords
popular_categories = df['about_product'].value_counts().reset_index()
popular_categories.columns = ['Category', 'Occurence']

popular_categories = popular_categories.sort_values(by='Occurence', ascending=False)
popular_categories
```

```python
# 7.What are the most popular product reviews
popular_reviews = df[["review_title", "review_content",'rating']].sort_values(by="rating",ascending=False)
popular_reviews.head(11)
```

```python
# 8. What is the correlation between discounted_price and rating
corr = df.discounted_price.corr(df.rating)
print("Weak Positive correlation between them.")
print(f"Correlation between discounted price and rating is: {corr:.4f}")

```

```python
# 9.What are the Top 5 categories based on the highest ratings?
top_5_rating_cat = df[df.rating == 5.0]
top_5_cat = df.groupby("category")["rating"].count().reset_index()
top_5_cat = top_5_cat.sort_values(by="rating",ascending=False)
top_5_cat.head()
```

```python
# 10. Identify any potential areas for improvement or optimization based on the data analysis.
print(df.dtypes)
# df.drop("product_id",axis=1, inplace=True)
df.isnull().sum()
df.discounted_price = df.discounted_price.fillna(df.groupby("product_name")["actual_price"].transform("mean"))
df.rating.fillna(df.rating.mean(),inplace=True)
df.drop


df.rating_count = df.rating_count.str.replace(",","")
df.rating_count = df.rating_count.astype(float)
df.rating_count = df.rating_count.fillna(df.rating.mean())
df.isnull().sum()
```
