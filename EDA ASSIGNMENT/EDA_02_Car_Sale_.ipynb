```python
# 1.What is the average selling price of cars for each dealer, and how does it compare across different dealers?
df.rename(columns={"Price ($)" : "Price"}, inplace =True)

average_price_per_dealer = df.groupby("Dealer_Name")["Price"].mean()

overall_avg_price = df.Price.mean()

comparing_df = pd.DataFrame({
    "Average_price" : average_price_per_dealer,
    "Overall_price": average_price_per_dealer - overall_avg_price})
comparing_df = comparing_df.sort_values(by="Average_price", ascending=False)

comparing_df
```

```python
# 2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
# trends?
price_variation = df.groupby("Company")['Price']
max_var_company = (price_variation.max() - price_variation.min()).idxmax()
print(f"{max_var_company} has the highest variance among all car companies")
'''
insights:
    Mercedes car company has the highest variance in prices among all the companies. 
    This indicates that Mercedes offers a wide range of cars, from budget-friendly options 
    to high-end luxury models.
'''
print()
```

```python
# 3.What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?

print(df.groupby('Transmission')["Price"].describe())
print()

iqr_values = df.groupby("Transmission")["Price"].apply(lambda x: x.quantile(0.75) - x.quantile(0.25))
print(f"IQR for each Transmission Type \n{iqr_values}")

# Visualising using bloxplot:
plt.figure(figsize=(13, 5.5))
sns.boxplot(x=df.Price, y= df.Transmission)
plt.title("Distribution of car prices for each Transmission", fontsize=11, fontweight="bold")
print()
plt.show()
```

```python
# 4. What is the distribution of car prices across different regions
print("Statistical Summary of Distribution of prices in different Regions")
print(df.groupby("Dealer_Region")["Price"].describe())

# visualizing using histogram
plt.figure(figsize=(12, 5))
sns.boxplot(x="Dealer_Region", y="Price", data = df,)
plt.title("Distribution of prices across regions", fontweight="bold", fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# 5. What is the distribution of cars based on body styles?
df.rename(columns={"Body Style	":"Body Style"},inplace=True)
print("Statistical information of the distribution on car based on body styles")
print(df.groupby("Body Style")["Price"].describe())

plt.figure(figsize=(12, 5))
sns.boxplot(x="Body Style", y="Price", data=df)
plt.title("Distribution of cars based on body styles", fontsize=11, fontweight='bold' )

plt.tight_layout()
plt.show()
```

```python
# 6. How does the average selling price of cars vary by customer gender and annual income?
df.rename(columns={"Annual Income	":"Annual_Income"}, inplace=True)
average_price_with_gender_income = df.groupby(["Gender","Annual Income"])["Price"].mean().reset_index()
print(average_price_with_gender_income)

plt.figure(figsize=(12,5))
sns.kdeplot(x="Annual Income", y="Price", hue="Gender", data=df)
plt.show()
```

```python
# Q7. What is the distribution of car prices by region, and how does the number of cars sold vary by region
print()
print("Statistical Information about the distribution")
print(df.groupby("Dealer_Region")["Price"].describe())

sales_by_region = df['Dealer_Region'].value_counts().reset_index()
sales_by_region.columns = ["Dealer_Region", 'No of cars sold']
sales_by_region
# Visualising the distribution
plt.figure(figsize=(13, 6))
plt.subplot(1,2, 1)
sns.boxplot(x="Dealer_Region", y="Price", data=df)
plt.title("Distribution of Car price by reagion")
plt.xlabel("Dealer regions")

# Visualizing how does the number of cars sold varies by region
plt.subplot(1,2, 2)
sns.barplot(x="Dealer_Region", y ="No of cars sold", data=sales_by_region)
plt.title("Number of cars sold varies by region")
plt.xlabel("Dealer regions")
plt.tight_layout()
plt.show()
```

```python
# Q8.How does the average car price differ between cars with different engine sizes?
avg_price_by_engine = df.groupby('Engine')['Price'].mean()
print("Average Car price with different sizes")
print(avg_price_by_engine)


avg_price_by_engine.plot(kind="bar")
plt.title("Average Car Price with different Engine sizes", fontweight="bold", fontsize=11)
plt.xlabel("Engine Sizes")
plt.ylabel("Average_Price")
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```

```python
# 9.How do car prices vary based on the customer’s annual income bracket?

bins = [0, 500000, 1500000, 4000000, float('inf')]
labels = ['Low income', 'Middle income', 'Upper-Middle income', 'High income']
df['Income_Bracket'] = pd.cut(df['Annual Income'], bins=bins, labels=labels)


avg_prices = df.groupby('Income_Bracket')['Price'].mean().reset_index()

# Display the results
print(avg_prices)

```

```python
# # 10.What are the top 5 car models with the highest number of sales, and how does their price distribution look

sales_count = df.Model.value_counts().reset_index()
sales_count.columns = ['Model' ,  "Sales_count"]

top_5_models = sales_count.head()

print("Top 5 Car Models with the highest number of sales")
print(top_5_models)

top_5_df = df[df['Model'].isin(top_5_models['Model'])]



# # visualizing the distribution using boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(x="Model", y="Price", data=top_5_df, order=top_5_models["Model"])
plt.title("Distribution of Top 5 Car model sales in number")

plt.tight_layout()
plt.show()
```

```python
'''11.How does car price vary with engine size across different car colors, and which colors have the highest price
variation'''
# car price variation with engine size across different colors
plt.figure(figsize=(13, 7))
plt.subplot(1,2,1)
sns.scatterplot(x="Engine", y="Price", hue="Color", data=df, palette="Set1")
plt.title("Car Price vs Engine Size across Different Car Colors")
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.legend(title="Car Color")

# price variation across car colors
plt.subplot(1,2,2)
sns.boxplot(x="Color", y="Price", data=df, palette="Set2")
plt.title("Price Variation across Different Car Colors")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 3. To find the colors with the highest price variation
price_variation = df.groupby('Color')['Price'].std()  # Standard deviation as price variation
highest_variation_color = price_variation.idxmax()
highest_variation_value = price_variation.max()

print(f"Color with highest price variation: {highest_variation_color} with a variation of {highest_variation_value}")
```

```python
# 12.Is there any seasonal trend in car sales based on the date of sale
monthly_sales = df.groupby(['Year', 'Month']).size().reset_index(name='Sales')
plt.figure(figsize=(12, 7))
sns.lineplot(x='Month', y='Sales', hue='Year', data=monthly_sales, marker="o")
plt.title("Seasonal Trend in Car Sales Based on Date of Sale", fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Number of Sales")
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(True)
plt.legend(title="Year")
plt.show()
```

```python
'''13.How does the car price distribution change when considering different combinations of body style and
transmission type'''
plt.figure(figsize=(15, 8))
for column in df.columns:
    column.strip()
sns.boxplot(x="Body Style", y="Price", hue="Transmission", data=df)

plt.title("Car Price Distribution by Body Style and Transmission Type", fontweight="bold")
plt.xlabel("Body Style")
plt.ylabel("Price")
plt.xticks(rotation=90)
plt.grid(True)
plt.legend(title="Transmission Type")
plt.tight_layout()
plt.show()
```

```python
'''14.What is the correlation between car price, engine size, and annual income of customers, and how do these
features interact?'''
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df.Engine = label_encoder.fit_transform(df.Engine)
# print("Label Encoded are:\n", label_encoded)
correlation_matrix = df[['Price', 'Engine', 'Annual Income']].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Price, Engine Size, and Annual Income", fontweight="bold")
plt.show()

# Pair Plot for Interaction Visualisatio
sns.pairplot(df[['Price', 'Engine', 'Annual Income']], diag_kind='kde', palette="husl")
plt.show()
```

```python
'''15.How does the average car price vary across different car models and engine types?'''
avg_price = df.groupby(['Model', 'Engine'])['Price'].mean().reset_index()

# Visualise
plt.figure(figsize=(15, 7))
sns.barplot(x='Model', y='Price', hue='Engine', data=avg_price)

plt.title("Average Car Price Across Models and Engine Types")
plt.xticks(rotation=90, fontsize=7)  # Rotate x-axis labels for better visibility
plt.xlabel("Car Model")
plt.ylabel("Average Price")
plt.tight_layout()
plt.show()

```
